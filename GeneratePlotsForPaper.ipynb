```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

mapping = {
            # 'Baseline_1': 'results/Baseline_1',
            # 'Baseline_2': 'results/Baseline_2',
            'Model13': 'results/Model_Run13',
             'Model_Run14' : 'results/Model_Run14',
           }


keys_to_plot = [
    'avg_delay_CAV', 'avg_delay_ALL', 'avg_speed_CAV','avg_speed_AllButCAV','congestion(avg_occupancy_priorityLane)','avg_PMx_emission'  
]

fig, axes = plt.subplots(2,3, figsize=(10,8), dpi=300)
axes = axes.flatten()

data = []
for i, (label, _filename) in enumerate(mapping.items()):
    filename = f'{_filename}.csv'
    df = pd.read_csv(filename)
    df['baseline'] = label
    df['seed'] = range(len(df))
    data.append(df)
    super_df = pd.concat(data).reset_index()
    # agg = df.groupby('Episode_Step').mean()
for j, key in enumerate(keys_to_plot):
    sns.lineplot(data=super_df, hue='baseline', x="Episode_Step", estimator='median', errorbar=('pi', 50), y=key,ax=axes[j])


# axes[2].legend()




 
```
