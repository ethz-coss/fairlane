```python
import pandas as pd
import matplotlib.pyplot as plt

filename_baseline_1 = 'results\Baseline_1.csv'
filename_baseline_2 = 'results\Baseline_2.csv'
df_baseline_1 = pd.read_csv(filename_baseline_1)
df_baseline_2 = pd.read_csv(filename_baseline_2)

```

```python
x = df_baseline_1["avg_delay_CAV"]
x1 = df_baseline_2["avg_delay_CAV"]

plt.plot(x, label = "Baseline 1")
plt.plot(x1, label = "Baseline 2")
plt.xlabel("200 Steps")
plt.ylabel("avg_delay_CAV")
plt.legend()
```

```python
x = df_baseline_1["avg_delay_ALL"]
x1 = df_baseline_2["avg_delay_ALL"]

plt.plot(x, label = "Baseline 1")
plt.plot(x1, label = "Baseline 2")
plt.xlabel("200 Steps")
plt.ylabel("avg_delay_ALL")
plt.legend()
```

```python
x = df_baseline_1["congestion(avg_occupancy_priorityLane)"]
x1 = df_baseline_2["congestion(avg_occupancy_priorityLane)"]

plt.plot(x, label = "Baseline 1")
plt.plot(x1, label = "Baseline 2")
plt.xlabel("200 Steps")
plt.ylabel("avg_occupancy_priorityLane")
plt.legend()
```

```python
x = df_baseline_1["avg_PMx_emission"]
x1 = df_baseline_2["avg_PMx_emission"]

plt.plot(x, label = "Baseline 1")
plt.plot(x1, label = "Baseline 2")
plt.xlabel("200 Steps")
plt.ylabel("avg_PMx_emission")
plt.legend()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

mapping = {
            'Baseline_1': 'results/Baseline_1',
            'Baseline_2': 'results/Baseline_2',
           }


keys_to_plot = [
    'avg_delay_CAV', 'avg_delay_ALL', 'congestion(avg_occupancy_priorityLane)','avg_PMx_emission'  
]

fig, axes = plt.subplots(2,2, figsize=(10,8), dpi=300)
axes = axes.flatten()
for i, (label, _filename) in enumerate(mapping.items()):
    filename = f'{_filename}.csv'
    df = pd.read_csv(filename)
    for j, key in enumerate(keys_to_plot):
        axes[j].plot(df[key], label=label)
        if i==0:
            axes[j].set_title(key)
axes[2].legend()
```
